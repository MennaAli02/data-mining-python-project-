```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# read the CSV file and select certain columns
df2 = pd.read_csv('C:\\Users\\hp\\OneDrive\\Desktop\\FCDS\\Level 2 Term 2\\Data Mining and Analytics\\Project\\imdb_movie_data.csv', usecols=['imdb_rating', 'vote'])

# compute the similarity scores between all pairs of records
similarity_matrix = cosine_similarity(df2)

# print the similarity scores
for i in range(len(similarity_matrix)):
    for j in range(i+1, len(similarity_matrix)):
        similarity_score = similarity_matrix[i][j]
        print(f"Similarity score between records {i+1} and {j+1}: {similarity_score}")
```
